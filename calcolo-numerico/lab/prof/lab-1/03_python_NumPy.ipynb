# **Tips on Numpy**

NumPy is a library for the Python programming language whose main objects are large and multi-dimensional arrays, which includes a large collection of high-level mathematical functions to operate on these arrays.

```
import numpy as np
```

### **DEFINING ARRAYS**

```
A=np.array([[1,2,3],[4,5,6]],dtype=np.float64) #a list becomes an array #we will see some exemples for which the choice of the dtype may affect the result
B=np.array((1,2,3,4)) #a tuple becomes an array
C= np.arange(9)
print(A,'\n')
print(B,'\n')
print(C,'\n')
```

```
print(A.size)
print(A.shape)
print(B.shape)
print(A.nbytes)
print(A.ndim)
print(A.dtype)

print(type(A.shape))
```

```
M=np.array([A,A,A,A])
print(M.shape)
print(M ,'\n')

N=np.zeros((2,3,3))
print(N,'\n')
N[:,:,0]=A

print(N.shape)
print(N)
```

```
print(np.diag(B))
```

```
print(np.zeros((3,3)))
```

```
print(np.ones((3,3)))
```

```
print(np.eye(3))
```

```
#Equidistant
D=np.linspace(0,10,num=10)
print(D)
```

```
#From a callable function
f = lambda m,n: n + 10*m
E = np.fromfunction(f, (3,3), dtype=np.float64)
print(E)
```

### **MATRIX-VECTOR MOLTIPLICATION**

```
A=[[1,2,3],[4,5,6],[7,8,9]]
A=np.array(A)
print(A)
print(A.shape)
```

```
v=np.ones((A.shape[1],1))
print(v)
print(v.shape)
```

```
result=np.matmul(A,v)
print(result)
print(result.shape)

```

```
result= A*v #watch out! This is not matrix-vector moltiplication
print(result)
```

```
x = np.ones((5,1))
y = np.linspace(1,5,5)
y = y.reshape((1,5))

print(np.dot(y,x))
print(np.matmul(y,x))
```

### **ARRAY MANIPULATION**

```
A=np.random.rand(2,2)
print(A)
```

```
B=np.transpose(A)
print(B,'\n')
C=A.T
print(B)
print(A==B)
print(C==B)
```

```
print(np.linalg.matrix_rank(A))
C=np.linalg.inv(A)
print(np.matmul(C,A),'\n')
print(C*A,'\n')
print(C@A)
```

### **AGGREGATION OF ARRAYS AND CONDITIONAL EXPRESSION**

```
t=np.sum(A)
print(t)
```

```
t = np.sum(A, axis=0)
print(t)
```

```
t = np.sum(A, axis=1)
print(t)
```

```
t=np.mean(A)
print(t)
```

```
t=np.std(A)
print(t)
```

```
t=np.prod(A)
print(t)
```

```
x=np.linspace(-10,10,21)
print(x,'\n')
print(x<0)

s=np.array(x[np.nonzero(x<0)])
print(s)

help(np.nonzero)
```

**FOR inside an array**

```
a=np.zeros((1,10))
print(a)

for i in range(np.size(a)):
  a[0,i]=1

print(a)
```

```
a=np.arange(30)
print(a,'\n')
a=np.reshape(a,(3,10))

print(a,'\n')

t = a[:,2]
print(t,'\n')

t = a[1,:]
print(t,'\n')

t = a[1:3,1:6]
print(t,'\n')
```
